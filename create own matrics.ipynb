```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
load_breast_cancer()
```

```python
#malignant-->0,benign-->1 convert to malignant-->1,bengin-->0
```

```python
x=load_breast_cancer().data
x=pd.DataFrame(x,columns=load_breast_cancer().feature_names)
x
```

```python
#changing the labels now
y=pd.Series(load_breast_cancer().target).map({0:1,1:0})
y
```

```python
#split the data train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#now we use ranfomforest
from sklearn.ensemble import RandomForestClassifier
Classifier=RandomForestClassifier(n_estimators=50,criterion='gini',max_depth=1)
```

```python
from sklearn.metrics import confusion_matrix,make_scorer
def fnr(y_true,y_pred):
    tn,fp,fn,tp=confussion_matrix(y_true,y_pred).ravel()
    return(fn)/(fn+tn)

fnr_scorer=make_scorer(fnr,greater_is_better=False)
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
#false negative scoring-fnrto apply an hyperparameter tuning
from sklearn.model_selection import GridSearchCV
parameters={'n_estimators':[10,20,30,40,50],
            'criterion':['gini','entorpy'],
            'max_depth':[1,2,3,4,5]
           }
gs=GridSearchCV(Classifier,param_grid=parameters,scoring=fnr_scorer,cv=5)
gs.fit(x_train,y_train)
```
